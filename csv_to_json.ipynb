```python
import csv
import json

def convert_csv_to_json(fileName):

    with open(fileName, newline='') as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        line_count = 0

        data = {}

        for row in csv_reader:
            if line_count == 0:
#                 print(f'Column names are {", ".join(row)}')
                line_count += 1
            else:

                # filename+size
                # fileref:''
                # size
                # filename
                # base64_img_data:''
                # file_attributes:{}
                # regions:{"0": {"shape_attributes":{name,x,y,width,height}, "region_attributes":{"Damage":"1"}}}

                file_dict = {}
                file_dict["fileref"] = ""
                file_dict["size"] = int(row[1])
                file_dict["filename"] = row[0]
                file_dict["base64_img_data"] = ""
                file_dict["file_attributes"] = json.loads(row[2])

                fileNameSize = row[0] + row[1]

                if int(row[4])>0:
                    regions_dict = data[fileNameSize]["regions"]
                else:
                    regions_dict = {}
                tmp_dict = {}
                shape_attribute_dict = json.loads(row[5])
                region_attribute_dict = json.loads(row[6])
                tmp_dict["shape_attributes"] = shape_attribute_dict
                tmp_dict["region_attributes"] = region_attribute_dict

                if len(shape_attribute_dict) == 0:
                    regions_dict = {}
                else:
                    regions_dict[row[4]] = tmp_dict

                file_dict["regions"] = regions_dict
                data[fileNameSize] = file_dict


    originalFileName, fileFormat = fileName.rsplit('.',1)

    with open(originalFileName + ".json", 'w', encoding='utf-8') as jsonf:
        jsonf.write(json.dumps(data, indent=4))


# Example:
convert_csv_to_json('via_region_data.csv')
```
